```python
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/gdrive/My Drive/pildoras/dataset/pastillitas/dataset/")
trainer.setTrainConfig(object_names_array=["pastilla"], batch_size=16, num_experiments=10, train_from_pretrained_model="/gdrive/My Drive/pildoras/dataset/pastillitas/pretrained-yolov3.h5")
trainer.trainModel()
```

```python
from imageai.Detection.Custom import DetectionModelTrainer

trainer = DetectionModelTrainer()
trainer.setModelTypeAsYOLOv3()
trainer.setDataDirectory(data_directory="/gdrive/My Drive/pildoras/dataset/pastillitas/dataset/")
metrics = trainer.evaluateModel(model_path="/gdrive/My Drive/pildoras/dataset/pastillitas/dataset/models", 
                                json_path="/gdrive/My Drive/pildoras/dataset/pastillitas/dataset/json/detection_config.json", 
                                iou_threshold=0.5, object_threshold=0.9, 
                                nms_threshold=0.5)
```

```python
from imageai.Detection.Custom import CustomObjectDetection

detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("/gdrive/My Drive/pildoras/dataset/pastillitas/dataset/models/detection_model-ex-010--loss-0026.422.h5")
detector.setJsonPath("/gdrive/My Drive/pildoras/dataset/pastillitas/dataset/json/detection_config.json")
detector.loadModel()
detections = detector.detectObjectsFromImage(input_image="/gdrive/My Drive/pildoras/dataset/aug1_img_45.jpg", 
                                             output_image_path="/gdrive/My Drive/pildoras/dataset/ima-detected3.jpg", 
                                             minimum_percentage_probability=30)
```
